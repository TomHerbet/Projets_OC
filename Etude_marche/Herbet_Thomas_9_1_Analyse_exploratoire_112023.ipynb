## ETAPE 3 : Extension PESTEL

```python
#Importation du fichier PIB.csv
PIB_csv ='/content/drive/MyDrive/Herbet_Thomas_P9/PIB.csv'
pib = pd.read_csv(PIB_csv, sep=',')
```

```python
pib.head()
```

```python
pib = pib.loc[(pib['Élément'].apply(lambda x: 'Valeur US $ par habitant' in x)) & (pib['Produit'].apply(lambda x: 'Revenu national brut' in x))]
pib
```

```python
#Filtrage sur l'année 2017 et conservation des colonnes 'Zone' et 'Valeur'
pib_2017 = pib.loc[pib['Année']==2017]
pib_2017 = pib_2017.loc[:,['Zone','Valeur']]
pib_2017.rename(columns={'Valeur':'RNB_habitant'},inplace=True)
pib_2017.head()
```

```python
#Importation du fichier poulet.csv
poulet_csv ='/content/drive/MyDrive/Herbet_Thomas_P9/poulet.csv'
poulet_df = pd.read_csv(poulet_csv, sep=',')
```

```python
#Filtrage sur l'année 2017
poulet_df = poulet_df.loc[poulet_df['Année']==2017]
poulet_df.head()
```

```python
importations_poulet_df = poulet_df.loc[poulet_df['Élément'].str.contains('Importations')]
importations_poulet_df
```

```python
#Conservation des colonnes 'Zone', 'Unité' et 'Valeur'
importations_poulet_df = importations_poulet_df.loc[:,['Zone','Unité','Valeur']]
importations_poulet_df.head()
```

```python
# Conversion
importations_poulet_df = importations_poulet_df.loc[importations_poulet_df['Unité']=='1000 US$']
importations_poulet_df = importations_poulet_df.loc[:,['Zone','Valeur']]
importations_poulet_df['Prix_poulet_US$'] = importations_poulet_df['Valeur']*1000
importations_poulet_df = importations_poulet_df.groupby('Zone').sum('Prix_poulet_US$')
importations_poulet_df.head()
```

```python
importations_poulet_df.loc[importations_poulet_df.duplicated()]
```

```python
#Importation du fichier production_poulet.csv
production_poulet_csv ='/content/drive/MyDrive/Herbet_Thomas_P9/production_poulet.csv'
production_poulet_df = pd.read_csv(production_poulet_csv, sep=',')
production_poulet_df
```

```python
#Filtrage sur l'année 2017
production_poulet_df = production_poulet_df.loc[production_poulet_df['Année']==2017]
```

```python
# Conversion
production_poulet_df = production_poulet_df.loc[:,['Zone','Valeur']]
production_poulet_df.rename(columns={ 'Valeur': 'nb_poulet_tonnes'}, inplace=True)
production_poulet_df['nb_poulet_tonnes'] = production_poulet_df['nb_poulet_tonnes']*1000
production_poulet_df.head()
```

```python
#Viande totale consommée
conso_viande_df = dispo_alim_viande.loc[:,['Produit','Zone','Nourriture']]
conso_viande_df = conso_viande_df.groupby('Zone').sum().reset_index()
conso_viande_df.rename(columns={'Nourriture':'Conso_viande_tonnes'},inplace=True)
conso_viande_df['Conso_viande_tonnes'] = conso_viande_df['Conso_viande_tonnes']*1000
conso_viande_df.head()
```

```python
#Poulet consommé
conso_poulet_df = dispo_alim_volaille
conso_poulet_df = conso_poulet_df.loc[:,['Zone','Nourriture']]
conso_poulet_df = conso_poulet_df.groupby('Zone').sum().reset_index()
conso_poulet_df.rename(columns={'Nourriture':'Conso_poulet_tonnes'},inplace=True)
conso_poulet_df['Conso_poulet_tonnes'] = conso_poulet_df['Conso_poulet_tonnes']*1000
conso_poulet_df.head()
```

Part de poulet consommé par rapport aux autres viandes

```python
part_poulet_viande_df = pd.merge(conso_poulet_df,conso_viande_df,how='inner',on='Zone')
part_poulet_viande_df['%poulet_consommé_viande'] = (part_poulet_viande_df['Conso_poulet_tonnes'])/(part_poulet_viande_df['Conso_viande_tonnes'])*100
part_poulet_viande_df = part_poulet_viande_df.loc[:,['Zone','%poulet_consommé_viande']]
part_poulet_viande_df.head()
```

Croissance démographique entre 2016 et 2017

```python
croissance_population_df = population_2016.merge(population_2017,how='left',on='Zone')
croissance_population_df['%croissance'] = ((croissance_population_df['nb_hab_17']-croissance_population_df['nb_hab_16'])/croissance_population_df['nb_hab_16'])*100
croissance_population_df = croissance_population_df.loc[:,['Zone','%croissance']]

croissance_population_df.head()
```

## ETAPE 3 : JOINTURE

```python
merged_df = pd.merge(dispo_alim_volaille, croissance_population_df, how='left', on='Zone')
merged_df = pd.merge(merged_df, population_2017, how='left', on='Zone')
merged_df = pd.merge(merged_df, pib_2017, how='left', on='Zone')
merged_df = pd.merge(merged_df, importations_poulet_df, how='left', on='Zone')
merged_df = pd.merge(merged_df, production_poulet_df, how='left', on='Zone')
merged_df = pd.merge(merged_df, conso_poulet_df, how='left', on='Zone')
merged_df = pd.merge(merged_df, conso_viande_df, how='left', on='Zone')
merged_df = pd.merge(merged_df, part_poulet_viande_df, how='left', on='Zone')
```

```python
merged_df
```

```python
merged_df['Tx_dépendance_importations'] = merged_df['Importations - Quantité']/(merged_df['Production']+merged_df['Importations - Quantité']-merged_df['Exportations - Quantité'])*100
```

```python
merged_df.drop(['Valeur','Conso_viande_tonnes','Disponibilité alimentaire (Kcal/personne/jour)','Disponibilité alimentaire en quantité (kg/personne/an)','Disponibilité de matière grasse en quantité (g/personne/jour)', 'Disponibilité de protéines en quantité (g/personne/jour)', 'Disponibilité intérieure', 'Nourriture', 'Variation de stock'],  axis=1, inplace=True)
```
